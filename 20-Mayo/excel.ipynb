 ## try/except
 En la programación, un error (también conocido como excepción) es un evento que ocurre durante la ejecución de un progra    ma cuando el programa encuentra una situación que no puede manejar.

Python utiliza lo que se conoce como excepciones para manejar errores. Cuando ocurre un error, Python crea un objeto especial llamado excepción. Si el código no maneja esta excepción, Python la propagará hasta el nivel superior del programa y, finalmente, detendrá la ejecución del programa.



```python
def leer_archivo(nombre_archivo):
    try:
        with open(nombre_archivo, 'r') as f:
            print(f.read())
    except FileNotFoundError:
        print(f"El archivo {nombre_archivo} no se encontró.")
    except PermissionError:
        print(f"No tienes permiso para leer el archivo {nombre_archivo}.")
    except Exception as e:
        print(f"Ocurrió un error inesperado al leer el archivo: {e}")

# Llamamos a la función con el nombre del archivo que queremos leer
leer_archivo("mi_archivo.txt")

# Ojo los errores FIleNotFoundError y PermissionError son subclases de OSError
# Por lo que podemos capturarlos con un solo except
# https://docs.python.org/3/library/exceptions.html#exception-hierarchy
```

# Packetes en python
En Python, un paquete es una forma de organizar módulos relacionados en un directorio. Un módulo es un archivo que contiene definiciones y declaraciones de Python. El nombre del módulo es el nombre del archivo.

Un paquete es simplemente un directorio que contiene un archivo especial llamado __init__.py (que puede estar vacío) y uno o más módulos. Aquí hay una representación simple de cómo se vería un paquete:

```python
import paquete.lector as lector
# Ejemplo de uso
ruta_archivo = "data/data.xlsx"

# Crear objeto de la clase Lector
lector_excel = lector.Lector(ruta_archivo)

# Obtener encabezados
encabezados = lector_excel.leer_encabezados()
print("Encabezados:")
print(encabezados)

# Obtener datos
datos = lector_excel.leer_excel()
print("Datos:")
print(datos)
```

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.



```python
!pip install matplotlib
```

```python
import matplotlib.pyplot as plt
plt.hist(datos['Edad'], bins=range(min(datos['Edad']), max(datos['Edad']) + 1), edgecolor='black')
plt.title("Distribución de edades")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Asegúrate de que los datos de experiencia sean numéricos
datos['Experiencia'] = datos['Experiencia'].str.extract('(\d+)').astype(int)

plt.figure(figsize=(10,6))
plt.title('Edad vs Experiencia')
plt.scatter(datos['Edad'], datos['Experiencia'], alpha=0.5)
plt.xlabel('Edad')
plt.ylabel('Experiencia (años)')
plt.show()
```

```python
!pip install seaborn

import seaborn as sns

plt.figure(figsize=(10,6))
sns.boxplot(x='Género', y='Edad', hue='Ciudad', data=datos, palette="Set1")
plt.title('Distribución de Edad por Género y Ciudad')
plt.show()
```
